# **Correlation** **Analysis**

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

relevant_columns = ['Temperature (째C)', 'Light_Intensity (lux)', 'Plant_Growth (cm)']
data_relevant = data[relevant_columns]

correlation_matrix = data_relevant.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title("Correlation Matrix: Temperature, Light Intensity, and Plant Growth")
plt.show()

print(correlation_matrix)
```

**Interpretation**:

The analysis revealed significant relationships between temperature, light intensity, and plant growth. Both temperature and light intensity are critical factors influencing plant growth, with temperature having a slightly stronger correlation.

1.A very strong positive correlation ( r=0.97) indicates that higher temperatures are strongly associated with increased plant growth.

2.Similarly, a strong positive correlation ( r=0.96) shows that greater light intensity promotes plant growth.

3.The highest correlation ( r=0.97) between these variables suggests that warmer days tend to have higher light intensity.

```python
correlation_with_growth = correlation_matrix['Plant_Growth (cm)'].drop('Plant_Growth (cm)')


strongest_factor = correlation_with_growth.idxmax()
strongest_correlation = correlation_with_growth.max()


print("Correlation with Plant Growth:")
print(correlation_with_growth)
print(f"\nThe environmental factor with the strongest correlation with plant growth is: {strongest_factor}")
print(f"Correlation coefficient: {strongest_correlation:.4f}")
```

 **Interpretation**:

1.Temperature has the strongest correlation with plant growth (r=0.97).

2.This suggests that variations in temperature are slightly more influential on plant growth compared to light intensity in this dataset

# **Linear** **Regression**

```python
#Linear Regression

import statsmodels.api as sm

X = data[['Temperature (째C)', 'Light_Intensity (lux)']]
y = data['Plant_Growth (cm)']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

**Interpretation**:

The linear regression analysis confirms that both temperature and light intensity significantly contribute to plant growth. Among the two, temperature has a slightly stronger influence. The high R square value indicates that the model is effective in explaining plant growth variability.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

X_temp = data[['Temperature (째C)']].values
X_light = data[['Light_Intensity (lux)']].values
y = data['Plant_Growth (cm)'].values

model_temp = LinearRegression().fit(X_temp, y)
model_light = LinearRegression().fit(X_light, y)

y_pred_temp = model_temp.predict(X_temp)
y_pred_light = model_light.predict(X_light)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x=X_temp.flatten(), y=y, color='blue', label='Actual Data')
plt.plot(X_temp.flatten(), y_pred_temp, color='red', label='Best-Fit Line')
plt.title('Temperature vs. Plant Growth')
plt.xlabel('Temperature (째C)')
plt.ylabel('Plant Growth (cm)')
plt.legend()

plt.subplot(1, 2, 2)
sns.scatterplot(x=X_light.flatten(), y=y, color='green', label='Actual Data')
plt.plot(X_light.flatten(), y_pred_light, color='orange', label='Best-Fit Line')
plt.title('Light Intensity vs. Plant Growth')
plt.xlabel('Light Intensity (lux)')
plt.ylabel('Plant Growth (cm)')
plt.legend()

plt.tight_layout()
plt.show()
```

**Interpretation**:

The visualizations demonstrate clear linear relationships between plant growth and the two environmental factors:

1.Temperature has a slightly stronger positive impact on plant growth.

2.Light Intensity also significantly contributes to plant growth, albeit to a lesser degree than temperature.

# **Trend** **Analysis**

```python

import matplotlib.pyplot as plt


data['Date'] = pd.to_datetime(data['Date'])

plt.figure(figsize=(12, 8))

plt.plot(data['Date'], data['Plant_Growth (cm)'], label='Plant Growth (cm)', color='green', linewidth=2)

plt.plot(data['Date'], data['Temperature (째C)'], label='Temperature (째C)', color='blue', linestyle='--')

plt.plot(data['Date'], data['Light_Intensity (lux)'], label='Light Intensity (lux)', color='orange', linestyle=':')

plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series of Plant Growth, Temperature, and Light Intensity')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

**Interpretation**:

The time series graph reveals clear temporal relationships between plant growth, temperature, and light intensity. It highlights the following:

1.Plant growth is influenced by seasonal variations in temperature and light intensity.

2.Higher values of temperature and light intensity generally correspond to increased plant growth.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

decomposition = seasonal_decompose(data['Plant_Growth (cm)'], model='additive', period=12)

plt.figure(figsize=(12, 10))
decomposition.plot()
plt.suptitle('Seasonal Decomposition of Plant Growth', fontsize=16)
plt.tight_layout()
plt.show()
```

**Interpretation**:

The analysis revealed distinct seasonal patterns and long-term trends in the data:

1.Plant growth is highly influenced by seasonal fluctuations in temperature and light intensity.

2.Long-term trends show a consistent increase in plant growth, likely due to favorable environmental conditions or controlled agricultural practices.
